```python
#Pohranjivanje podataka u dataframe
df_btc = pd.read_csv('BTC_kn_2021-01-03_2022-01-02-CoinDesk.csv')
df_eth = pd.read_csv('ETH_kn_2021-01-03_2022-01-02-CoinDesk.csv')
df_ltc = pd.read_csv('LTC_kn_2021-01-03_2022-01-02-CoinDesk.csv')
df_xrp = pd.read_csv('XRP_kn_2021-01-01_2022-01-02-CoinDesk.csv')
```

```python
#Kreiranje novog dataframea koji sadrži završne cijene sve 4 kriptovalute
df = pd.DataFrame({'BTC': df_btc['Closing Price (USD)'],
                   'ETH': df_eth['Closing Price (USD)'],
                   'LTC': df_ltc['Closing Price (USD)'],
                   'XRP': df_xrp['Closing Price (USD)']
})      
```

```python
#Pokazivanje novog dataframea
df
```

```python
#Statistički podaci
df.describe()
```

```python
#Vizualizacija kriptokurentskih zaključnih cijena
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

my_crypto = df
plt.figure(figsize = (12.2, 4.5))
for c in my_crypto.columns.values:
  plt.plot(my_crypto[c], label = c)

plt.title('Graf kriptovaluta')
plt.xlabel('Dani')
plt.ylabel('Cijena kriptovalute u $')
plt.legend(my_crypto.columns.values, loc= 'upper left')
plt.show()
```

```python
#Skaliranje podataka
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled = min_max_scaler.fit_transform(df)
scaled
```

```python
#pretvorba skaliranih podataka u okvir podataka
df_scale = pd.DataFrame(scaled, columns= df.columns)
```

```python
#Vizualizacija skaliranih podataka
my_crypto = df_scale

plt.figure(figsize=(12.4, 4.5))
for c in my_crypto.columns.values:
  plt.plot(my_crypto[c], label=c)

plt.title('Vizualizacija skaliranih podataka')
plt.xlabel('Dani')
plt.ylabel('Kripto skalirani podaci u $')
plt.legend(my_crypto.columns.values, loc='upper left')
plt.show()
```

```python
#jednostavni dnevni povratak
DSR = df.pct_change(1)
DSR
```

```python
#Vizualizacija jednostavnog dnevnog povratka
plt.figure(figsize=(12.4,4.5))

for c in DSR.columns.values:
  plt.plot(DSR.index, DSR[c], label = c, lw = 2, alpha = .7)

plt.title('Jednostavni dnevni povratak')
plt.ylabel('Postotci (u decimalnoj formi)')
plt.xlabel('Dani')
plt.legend(DSR.columns.values, loc='upper right')
plt.show()
```

```python
#Volatilnost dobiti
print('Volatilnost kriptovalute')
DSR.std()
```

```python
#Pokazuje aritmetičku sredinu dnevnog povrata
DSR.mean()
```

```python
#korelacija
DSR.corr()
```

```python
#Vizualizacija korelacije
import seaborn as sns

plt.subplots(figsize=(11,11))
sns.heatmap(DSR.corr(), annot = True, fmt=' .2%')
```

```python
#Dnevni kumulativni povrat
DCSR = (DSR+1).cumprod()

#Kako to izgleda?
DCSR
```

```python
#Vizualizacija dnevnog kumulativnog povrata
plt.figure(figsize=(12.2, 4.5))
for c in DCSR.columns.values:
  plt.plot(DCSR.index, DCSR[c], lw=2, label=c)

plt.title('Dnevni kumulativni povrat')
plt.xlabel('Dani')
plt.ylabel('Koliki je dobitah odnosno gubitak na ulog od 1$')
plt.legend(DCSR.columns.values, loc = 'upper left', fontsize = 10)
plt.show()
```
